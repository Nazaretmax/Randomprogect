# 📱 HTML2APK Universal — Cordova Builder

> Converte qualsiasi progetto HTML/JS/CSS in APK Android firmato — pronto per Google Play Store

---

## 🗺️ Celle in ordine:

| # | Nome | Cosa fa |
|---|------|---------|
| **⚙️ CONFIG** | Configurazione | **✏️ Modifica qui: nome, package, versione, URL o file** |
| **1** | Setup | Installa Java 17, Android SDK, Node.js, Gradle, Cordova |
| **2** | Risorse | Carica icona + file HTML (se usi file locali) |
| **3** | Progetto | Crea progetto Cordova + bundle CDN offline |
| **4** | Permessi | Configura permessi app |
| **5** | Keystore | Genera/carica chiave per firma (Google Play) |
| **6** | Build | Compila APK debug + release |
| **7** | Download | Scarica APK |
| **🔍 DEBUG** | Debug | Log errori |

```
# ╔══════════════════════════════════════════════════════════════╗
# ║          ⚙️  CONFIGURAZIONE — MODIFICA QUI                  ║
# ╚══════════════════════════════════════════════════════════════╝

# ── Informazioni App ──────────────────────────────────────────
APP_NAME        = 'Ninjarevers'            # Nome visualizzato sull'icona
PKG_NAME        = 'com.ninjarevers.game'   # Package univoco (com.tuonome.app)
APP_VERSION     = '1.0.0'                  # Versione (es: 1.0.0, 2.1.3)
APP_VERSION_CODE = 1                       # Numero intero crescente (1, 2, 3...)
APP_DESCRIPTION = 'Ninja Revers Game'      # Descrizione breve
APP_AUTHOR      = 'Dev'                    # Nome autore
APP_AUTHOR_EMAIL= 'dev@example.com'        # Email autore

# ── Sorgente contenuto ────────────────────────────────────────
# Scegli UNO dei due metodi:
#
# METODO A — URL (app carica un sito web, richiede internet)
#   SOURCE_TYPE = 'url'
#   SOURCE_URL  = 'https://il-tuo-sito.com'
#
# METODO B — File locali (HTML+JS+CSS nell'APK, funziona offline)
#   SOURCE_TYPE = 'files'
#   SOURCE_URL  = ''   (ignorato)

SOURCE_TYPE = 'files'                      # 'url' oppure 'files'
SOURCE_URL  = 'https://example.com'        # URL solo se SOURCE_TYPE='url'

# ── Orientamento ─────────────────────────────────────────────
# 'portrait'  → solo verticale
# 'landscape' → solo orizzontale
# 'default'   → segue il dispositivo
ORIENTATION = 'portrait'

# ── Versione Android target ───────────────────────────────────
MIN_SDK_VERSION    = 24     # Android 7.0 minimo (24=7.0, 26=8.0, 28=9.0)
TARGET_SDK_VERSION = 34     # Android 14 target (consigliato per Play Store 2024)

# ── Permessi ─────────────────────────────────────────────────
PERM_INTERNET       = True   # Accesso internet (necessario per CDN/API)
PERM_VIBRATE        = True   # Vibrazione (utile per giochi)
PERM_AUDIO          = True   # AudioContext / Web Audio API
PERM_CAMERA         = False  # Accesso fotocamera
PERM_MICROPHONE     = False  # Accesso microfono
PERM_GEOLOCATION    = False  # GPS / posizione
PERM_STORAGE_READ   = True   # Leggi storage esterno
PERM_STORAGE_WRITE  = True   # Scrivi storage esterno

# ── WebView avanzate ─────────────────────────────────────────
WEBVIEW_JS              = True   # JavaScript abilitato
WEBVIEW_DOM_STORAGE     = True   # localStorage / sessionStorage
WEBVIEW_DATABASE        = True   # IndexedDB / Web SQL
WEBVIEW_FILE_ACCESS     = True   # Accesso file locali
WEBVIEW_UNIVERSAL_ACCESS= True   # Cross-origin per file:// (CDN inline)
WEBVIEW_MIXED_CONTENT   = True   # HTTP+HTTPS insieme
WEBVIEW_MEDIA_AUTOPLAY  = True   # Audio/Video autoplay senza tap
WEBVIEW_FULLSCREEN      = True   # Nasconde barre sistema
WEBVIEW_HARDWARE_ACCEL  = True   # GPU acceleration (Canvas/WebGL)
WEBVIEW_KEEP_SCREEN_ON  = True   # Schermo sempre acceso (giochi)
WEBVIEW_CACHE_MODE      = 'default'  # 'default', 'no_cache', 'cache_first'

# ── Keystore (firma APK) ─────────────────────────────────────
# Per Google Play Store devi usare SEMPRE lo stesso keystore!
# Se hai già un keystore, impostare KEYSTORE_EXISTING = True e caricarlo
# Se è la prima volta, lascia KEYSTORE_EXISTING = False
KEYSTORE_EXISTING   = False          # True = carica esistente, False = genera nuovo
KEYSTORE_ALIAS      = 'appkey'       # Nome alias chiave
KEYSTORE_PASSWORD   = 'MyPass123!'   # Password keystore (cambiala!)
KEYSTORE_VALIDITY   = 10000          # Giorni validità (10000 ≈ 27 anni)
KEYSTORE_CN         = APP_NAME       # Common Name
KEYSTORE_ORG        = 'MyCompany'    # Organizzazione
KEYSTORE_COUNTRY    = 'IT'           # Codice paese (IT, US, GB...)

# ── Build type ────────────────────────────────────────────────
# 'debug'   → più veloce, per test
# 'release' → ottimizzato, per Google Play
# 'both'    → genera entrambi
BUILD_TYPE = 'both'

print('━'*60)
print('⚙️  CONFIGURAZIONE CARICATA')
print('━'*60)
print(f'📱 App:         {APP_NAME} v{APP_VERSION} ({APP_VERSION_CODE})')
print(f'📦 Package:     {PKG_NAME}')
print(f'🌐 Sorgente:    {SOURCE_TYPE.upper()}' + (f' → {SOURCE_URL}' if SOURCE_TYPE=='url' else ' → file locali'))
print(f'📐 Orient.:     {ORIENTATION}')
print(f'🤖 Android:     {MIN_SDK_VERSION} → {TARGET_SDK_VERSION}')
print(f'🔑 Keystore:    {"Esistente" if KEYSTORE_EXISTING else "Nuovo"}')
print(f'🔨 Build:       {BUILD_TYPE}')
print('━'*60)
print('✅ Configurazione OK → esegui Cella 1')
```

```
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELLA 1 — Setup ambiente (Java, Android SDK, Node, Gradle, Cordova)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
import os, subprocess, sys

ANDROID_HOME  = '/opt/android-sdk'
JAVA_HOME     = '/usr/lib/jvm/java-17-openjdk-amd64'
GRADLE_VER    = '7.6.4'
GRADLE_HOME   = f'/opt/gradle-{GRADLE_VER}'
SDK_MGR       = f'{ANDROID_HOME}/cmdline-tools/latest/bin/sdkmanager'

def set_env():
    os.environ['ANDROID_HOME'] = ANDROID_HOME
    os.environ['JAVA_HOME']    = JAVA_HOME
    os.environ['GRADLE_HOME']  = GRADLE_HOME
    os.environ['PATH'] = (
        f'{JAVA_HOME}/bin:'
        f'{GRADLE_HOME}/bin:'
        f'{ANDROID_HOME}/cmdline-tools/latest/bin:'
        f'{ANDROID_HOME}/platform-tools:'
        f'{ANDROID_HOME}/build-tools/33.0.2:'
        '/usr/local/bin:/usr/bin:/bin'
    )
set_env()

def run(cmd, **kw): return subprocess.run(cmd, capture_output=True, text=True, **kw)

# ── Java 17 ────────────────────────────────────────────────────
print('⏳ [1/5] Java 17...')
run(['apt-get','install','-y','openjdk-17-jdk-headless','wget','unzip'])
rv = run(['java','-version'])
print('  ✅', (rv.stderr or rv.stdout).split('\n')[0].strip())

# ── Android SDK ────────────────────────────────────────────────
print('⏳ [2/5] Android SDK...')
if not os.path.exists(f'{ANDROID_HOME}/build-tools/33.0.2/aapt'):
    run(['wget','-q',
         'https://dl.google.com/android/repository/commandlinetools-linux-9477386_latest.zip',
         '-O','/tmp/ct.zip'])
    run(['unzip','-q','/tmp/ct.zip','-d','/tmp/ct'])
    os.makedirs(f'{ANDROID_HOME}/cmdline-tools/latest', exist_ok=True)
    run(f'mv /tmp/ct/cmdline-tools/* {ANDROID_HOME}/cmdline-tools/latest/', shell=True)
    run(f'yes | {SDK_MGR} --licenses > /dev/null 2>&1', shell=True)
    run(f'yes | {SDK_MGR} "platforms;android-33" "platforms;android-34" "build-tools;33.0.2" "platform-tools"',
        shell=True)
print('  ✅ Android SDK (API 33+34, build-tools 33.0.2)')

# ── Gradle ─────────────────────────────────────────────────────
print('⏳ [3/5] Gradle...')
if not os.path.exists(f'{GRADLE_HOME}/bin/gradle'):
    run(['wget','-q',
         f'https://services.gradle.org/distributions/gradle-{GRADLE_VER}-bin.zip',
         '-O','/tmp/gradle.zip'])
    run(['unzip','-q','/tmp/gradle.zip','-d','/opt/'])
rv = run(['gradle','--version'])
gv = next((l for l in rv.stdout.split('\n') if 'Gradle' in l), '?')
print(f'  ✅ {gv.strip()}')

# ── Node.js 18 ─────────────────────────────────────────────────
print('⏳ [4/5] Node.js 18...')
nv = run(['node','--version'])
if nv.returncode != 0 or not nv.stdout.startswith('v18'):
    run('curl -fsSL https://deb.nodesource.com/setup_18.x | bash - > /dev/null 2>&1', shell=True)
    run(['apt-get','install','-y','nodejs'])
nv = run(['node','--version'])
npmv = run(['npm','--version'])
print(f'  ✅ Node {nv.stdout.strip()} / npm {npmv.stdout.strip()}')

# ── Cordova CLI ────────────────────────────────────────────────
print('⏳ [5/5] Cordova CLI...')
cv = run(['cordova','--version'])
if cv.returncode != 0:
    run(['npm','install','-g','cordova@12'])
cv = run(['cordova','--version'])
print(f'  ✅ Cordova {cv.stdout.strip()}')

# ── Python deps ────────────────────────────────────────────────
run([sys.executable,'-m','pip','install','requests','Pillow','-q'])

# ── Riepilogo ──────────────────────────────────────────────────
print()
print('━'*60)
checks = [
    ('Java 17',          os.path.exists(f'{JAVA_HOME}/bin/java')),
    ('Android SDK',      os.path.exists(f'{ANDROID_HOME}/build-tools/33.0.2/aapt')),
    ('Gradle',           os.path.exists(f'{GRADLE_HOME}/bin/gradle')),
    ('Node.js',          run(['node','--version']).returncode == 0),
    ('Cordova',          run(['cordova','--version']).returncode == 0),
]
all_ok = True
for name, ok in checks:
    print(f'  {"✅" if ok else "❌"} {name}')
    if not ok: all_ok = False
print('━'*60)
print(f'{"✅ Cella 1 OK → esegui Cella 2" if all_ok else "❌ Qualcosa manca — riprova"}')
```

```
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELLA 2 — Carica risorse (icona + file HTML se SOURCE_TYPE='files')
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
from google.colab import files as colab_files
import os

zip_file  = None
icon_file = None

# ── Icona ──────────────────────────────────────────────────────
print('='*60)
print('🖼️  Carica ICONA (PNG, almeno 512×512 consigliata)')
print('='*60)
up_icon = colab_files.upload()
for fname, data in up_icon.items():
    icon_file = f'/content/{fname}'
    open(icon_file,'wb').write(data)
    print(f'  ✅ Icona: {fname} ({len(data)/1024:.0f} KB)')

# ── File HTML (solo se SOURCE_TYPE = 'files') ──────────────────
if SOURCE_TYPE == 'files':
    print()
    print('='*60)
    print('📄 Carica ZIP progetto HTML')
    print('   Il ZIP deve contenere: index.html + JS/CSS/immagini')
    print('   Tutti i file devono stare nella ROOT dello ZIP')
    print('   (oppure in una cartella, verrà estratta automaticamente)')
    print('='*60)
    up_zip = colab_files.upload()
    for fname, data in up_zip.items():
        if fname.lower().endswith('.zip'):
            zip_file = f'/content/{fname}'
            open(zip_file,'wb').write(data)
            print(f'  ✅ ZIP: {fname} ({len(data)/1024:.0f} KB)')
        else:
            print(f'  ⚠️  {fname}: deve essere .zip')
else:
    print(f'\n  ℹ️  SOURCE_TYPE=url → nessun file da caricare')
    print(f'  🌐 URL: {SOURCE_URL}')

print()
ok = icon_file and (SOURCE_TYPE == 'url' or zip_file)
print('✅ Cella 2 OK → esegui Cella 3' if ok else '❌ Manca qualcosa')
```

```
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELLA 3 — Crea progetto Cordova + copia file + bundle CDN
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
import os, re, base64, zipfile, shutil, subprocess, requests
from PIL import Image

def set_env():
    os.environ.update({
        'ANDROID_HOME': '/opt/android-sdk',
        'JAVA_HOME':    '/usr/lib/jvm/java-17-openjdk-amd64',
        'GRADLE_HOME':  f'/opt/gradle-7.6.4',
    })
    os.environ['PATH'] = (
        '/usr/lib/jvm/java-17-openjdk-amd64/bin:'
        '/opt/gradle-7.6.4/bin:'
        '/opt/android-sdk/cmdline-tools/latest/bin:'
        '/opt/android-sdk/platform-tools:'
        '/opt/android-sdk/build-tools/33.0.2:'
        '/usr/local/bin:/usr/bin:/bin'
    )
set_env()

PROJ = f'/content/{APP_NAME}'
if os.path.exists(PROJ): shutil.rmtree(PROJ)

# ── 1. Crea progetto Cordova ────────────────────────────────────
print('⏳ [1/4] Creazione progetto Cordova...')
r = subprocess.run(['cordova','create', PROJ, PKG_NAME, APP_NAME],
                   capture_output=True, text=True)
if r.returncode != 0: print('❌', r.stderr[:300]); raise SystemExit()
print(f'  ✅ {PROJ}')

# ── 2. Aggiungi piattaforma Android ────────────────────────────
print('⏳ [2/4] Piattaforma Android...')
r2 = subprocess.run(['cordova','platform','add','android@12'],
                    capture_output=True, text=True, cwd=PROJ)
if r2.returncode != 0 and 'already' not in (r2.stderr+r2.stdout).lower():
    print('⚠️', r2.stderr[-300:])
else:
    print('  ✅ Android 12 platform')

# ── 3. Copia contenuto ─────────────────────────────────────────
print('⏳ [3/4] Contenuto app...')
www = os.path.join(PROJ, 'www')
shutil.rmtree(www); os.makedirs(www)

if SOURCE_TYPE == 'url':
    # Crea una pagina HTML che carica l'URL nella WebView
    open(os.path.join(www,'index.html'),'w').write(f'''<!DOCTYPE html>
<html><head>
<meta charset="UTF-8"/>
<meta name="viewport" content="width=device-width,initial-scale=1.0"/>
<style>*{{margin:0;padding:0;}}iframe{{width:100vw;height:100vh;border:none;}}</style>
</head><body>
<iframe src="{SOURCE_URL}" allowfullscreen></iframe>
<script src="cordova.js"></script>
</body></html>''')
    print(f'  ✅ WebView URL: {SOURCE_URL}')
else:
    # Estrai ZIP
    with zipfile.ZipFile(zip_file) as z:
        names = [n for n in z.namelist() if not n.endswith('/')]
        # Rileva cartella radice
        roots = set(n.split('/')[0] for n in z.namelist())
        single_root = len(roots)==1 and list(roots)[0]+'/' in z.namelist()
        root_prefix = list(roots)[0]+'/' if single_root else ''
        for name in names:
            target = name[len(root_prefix):] if root_prefix and name.startswith(root_prefix) else name
            if not target: continue
            dest = os.path.join(www, target)
            os.makedirs(os.path.dirname(dest), exist_ok=True)
            with z.open(name) as src, open(dest,'wb') as dst:
                dst.write(src.read())
    # Rinomina se serve
    htmls = [f for f in os.listdir(www) if f.endswith('.html')]
    if htmls and 'index.html' not in htmls:
        os.rename(os.path.join(www,htmls[0]), os.path.join(www,'index.html'))
        print(f'  ✅ Rinominato {htmls[0]} → index.html')
    n_files = sum(len(fs) for _,_,fs in os.walk(www))
    print(f'  ✅ {n_files} file estratti in www/')

    # ── Bundle CDN offline ──────────────────────────────────────
    print('  ⏳ Bundle CDN offline...')
    html_path = os.path.join(www,'index.html')
    with open(html_path, encoding='utf-8') as f: html = f.read()

    H = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    cdn_list = [
        ('https://cdn.tailwindcss.com',                                      'Tailwind'),
        ('https://unpkg.com/react@18/umd/react.production.min.js',          'React 18'),
        ('https://unpkg.com/react-dom@18/umd/react-dom.production.min.js',  'ReactDOM 18'),
        ('https://unpkg.com/@babel/standalone/babel.min.js',                'Babel'),
    ]
    cdn_map = {}
    for url, name in cdn_list:
        # Salta se non presente nell'HTML
        if url not in html:
            print(f'    ⏭  {name} (non usato)')
            continue
        try:
            r = requests.get(url, headers=H, timeout=90)
            cdn_map[url] = r.text
            print(f'    ✅ {name}: {len(r.text)//1024} KB')
        except Exception as e:
            print(f'    ❌ {name}: {e}')

    # Fonts Google
    font_link = re.search(r'<link[^>]+fonts\.googleapis\.com/css2[^>]*>', html)
    if font_link:
        try:
            font_url = re.search(r'href=["\']([^"\']+)["\']', font_link.group()).group(1)
            fc = requests.get(font_url, headers=H, timeout=20).text
            for fu in re.findall(r'url\((https://fonts\.gstatic[^)]+)\)', fc):
                try:
                    fd = requests.get(fu, timeout=10)
                    b64 = base64.b64encode(fd.content).decode()
                    mime = 'font/woff2' if 'woff2' in fu else 'font/woff'
                    fc = fc.replace(fu, f'data:{mime};base64,{b64}')
                except: pass
            html = re.sub(r'<link[^>]+fonts\.googleapis\.com[^>]*/?>\s*','',html)
            html = html.replace('<head>',f'<head>\n<style>\n{fc}\n</style>',1)
            print(f'    ✅ Google Fonts inline: {len(fc)//1024} KB')
        except Exception as e:
            print(f'    ⚠️  Fonts: {e}')

    # Inline CDN scripts
    for url, name in cdn_list:
        if url not in cdn_map: continue
        content = cdn_map[url]
        pat = re.compile(r'<script[^>]+src=["\']'+re.escape(url)+r'["\'][^>]*>\s*</script>', re.I)
        m = pat.search(html)
        new_tag = f'<script>\n{content}\n</script>'
        html = html.replace(m.group(0), new_tag, 1) if m else html.replace('</head>',new_tag+'\n</head>',1)
        print(f'    ✅ {name} inline')

    # Android/Cordova fix
    fix = '''<script src="cordova.js"></script>
<script>
document.addEventListener('deviceready',function(){
  // localStorage polyfill
  try{localStorage.setItem('t','1');localStorage.removeItem('t');}
  catch(e){var S={};Object.defineProperty(window,'localStorage',{get:function(){
    return{getItem:function(k){return S[k]||null;},setItem:function(k,v){S[k]=String(v);},
           removeItem:function(k){delete S[k];},clear:function(){S={};},
           get length(){return Object.keys(S).length;}};}})
  }
  // AudioContext
  var AC=window.AudioContext||window.webkitAudioContext;
  if(AC){var o=AC;window.AudioContext=window.webkitAudioContext=function(){var c=new o();c.resume&&c.resume();return c;}}
  // Blocca history React
  History.prototype.pushState=function(){};
  History.prototype.replaceState=function(){};
  // Blocca zoom doppio tap
  var lt=0;
  document.addEventListener('touchend',function(e){var n=Date.now();if(n-lt<350)e.preventDefault();lt=n;},{passive:false});
  document.addEventListener('touchmove',function(e){if(e.touches.length>1)e.preventDefault();},{passive:false});
  console.log('[OK] deviceready');
},false);
</script>'''
    html = html.replace('<head>','<head>\n'+fix,1)
    with open(html_path,'w',encoding='utf-8') as f: f.write(html)
    print(f'  ✅ HTML bundled: {os.path.getsize(html_path)/1024/1024:.1f} MB')

# ── 4. Icona ────────────────────────────────────────────────────
print('⏳ [4/4] Icone...')
res_dir = os.path.join(PROJ,'res')
os.makedirs(res_dir, exist_ok=True)
img = Image.open(icon_file).convert('RGBA').resize((1024,1024),Image.LANCZOS)
img.save(os.path.join(res_dir,'icon.png'),'PNG')
# Splash (sfondo nero)
splash = Image.new('RGBA',(2732,2732),(5,13,26,255))
splash.save(os.path.join(res_dir,'splash.png'),'PNG')
print('  ✅ Icona 1024×1024 + splash')

print()
print('━'*60)
print(f'✅ Progetto Cordova pronto: {PROJ}')
print('━'*60)
print('✅ Cella 3 OK → esegui Cella 4')
```

```
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELLA 6 — Build APK (debug e/o release)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
import os, subprocess, glob, shutil

PROJ = f'/content/{APP_NAME}'
BT   = '/opt/android-sdk/build-tools/33.0.2'

def set_env():
    os.environ.update({
        'ANDROID_HOME': '/opt/android-sdk',
        'JAVA_HOME':    '/usr/lib/jvm/java-17-openjdk-amd64',
        'GRADLE_HOME':  '/opt/gradle-7.6.4',
    })
    os.environ['PATH'] = (
        '/usr/lib/jvm/java-17-openjdk-amd64/bin:'
        '/opt/gradle-7.6.4/bin:'
        '/opt/android-sdk/cmdline-tools/latest/bin:'
        '/opt/android-sdk/platform-tools:'
        f'{BT}:/usr/local/bin:/usr/bin:/bin'
    )
set_env()

def sign_apk(src, out_name):
    aligned = f'/content/{out_name}_aligned.apk'
    signed  = f'/content/{out_name}.apk'
    # zipalign
    subprocess.run([f'{BT}/zipalign','-v','-f','4',src,aligned], capture_output=True)
    inp = aligned if os.path.exists(aligned) else src
    # apksigner
    r = subprocess.run([
        f'{BT}/apksigner','sign',
        '--ks', KEYSTORE_PATH,
        '--ks-key-alias', KEYSTORE_ALIAS,
        '--ks-pass',  f'pass:{KEYSTORE_PASSWORD}',
        '--key-pass', f'pass:{KEYSTORE_PASSWORD}',
        '--out', signed, inp
    ], capture_output=True, text=True)
    if r.returncode != 0:
        print(f'    ❌ Firma: {r.stderr[:300]}')
        return src
    # Verifica
    rv = subprocess.run([f'{BT}/apksigner','verify','--verbose',signed],
                        capture_output=True, text=True)
    ok = 'Verified using v' in rv.stdout
    print(f'    {"✅" if ok else "⚠️"} Firma verificata: {ok}')
    return signed

# ── Installa plugin ────────────────────────────────────────────
print('⏳ [1/3] Plugin Cordova...')
plugins = ['cordova-plugin-whitelist','cordova-plugin-file','cordova-plugin-media']
if PERM_GEOLOCATION: plugins.append('cordova-plugin-geolocation')
if PERM_CAMERA:      plugins.append('cordova-plugin-camera')
for p in plugins:
    r = subprocess.run(['cordova','plugin','add',p], capture_output=True, text=True, cwd=PROJ)
    print(f'  {"✅" if r.returncode==0 or "already" in (r.stdout+r.stderr).lower() else "⚠️"} {p}')

# ── Build ──────────────────────────────────────────────────────
results = {}
build_modes = []
if BUILD_TYPE in ('debug','both'):   build_modes.append('debug')
if BUILD_TYPE in ('release','both'): build_modes.append('release')

for mode in build_modes:
    print(f'\n⏳ [2/3] Build {mode.upper()} (3-5 min)...')
    r = subprocess.run(
        ['cordova','build','android',f'--{mode}'],
        capture_output=True, text=True, cwd=PROJ
    )
    if r.returncode != 0:
        print(f'  ❌ Build {mode} fallita:')
        print(r.stderr[-2000:])
        continue
    print(f'  ✅ Build {mode} completata')

    # Trova APK
    apk_files = glob.glob(f'{PROJ}/platforms/android/**/*.apk', recursive=True)
    apk_files = [a for a in apk_files if mode in a.lower()]
    if not apk_files:
        apk_files = glob.glob(f'{PROJ}/platforms/android/**/*.apk', recursive=True)
    if not apk_files:
        print(f'  ❌ APK non trovato'); continue

    src_apk = apk_files[0]
    print(f'  📦 APK: {src_apk} ({os.path.getsize(src_apk)/1024/1024:.1f} MB)')

    # Firma
    print(f'⏳ [3/3] Firma APK {mode}...')
    out_name = f'{APP_NAME}_v{APP_VERSION}_{mode}'
    signed = sign_apk(src_apk, out_name)
    results[mode] = signed
    print(f'  ✅ {os.path.basename(signed)} ({os.path.getsize(signed)/1024/1024:.1f} MB)')

# ── Riepilogo ──────────────────────────────────────────────────
print()
print('━'*60)
print('🎉 BUILD COMPLETATA')
for mode, path in results.items():
    print(f'  {"🐛" if mode=="debug" else "🚀"} {mode.upper()}: {os.path.basename(path)} ({os.path.getsize(path)/1024/1024:.1f} MB)')
print()
print('💡 Debug APK   → test sul telefono')
print('💡 Release APK → carica su Google Play Store')
print('━'*60)
print('✅ Cella 6 OK → esegui Cella 7 per scaricare')
```

```
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELLA DEBUG — Esegui solo se hai errori
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
import os, subprocess, glob

PROJ = f'/content/{APP_NAME}'

print('═'*60)
print('🔍 DEBUG REPORT')
print('═'*60)

# Versioni
print('\n──── Versioni ────')
for cmd, label in [(['java','-version'],'Java'),(['gradle','--version'],'Gradle'),
                   (['node','--version'],'Node'),(['cordova','--version'],'Cordova')]:
    r = subprocess.run(cmd, capture_output=True, text=True)
    out = (r.stdout or r.stderr).split('\n')[0].strip()
    print(f'  {label}: {out}')

# Env
print('\n──── Environment ────')
for v in ['ANDROID_HOME','JAVA_HOME','GRADLE_HOME']:
    print(f'  {v} = {os.environ.get(v,"❌ NON IMPOSTATO")}')

# APK
print('\n──── APK trovati ────')
found = False
for root, dirs, fnames in os.walk('/content'):
    for f in fnames:
        if f.endswith('.apk'):
            p = os.path.join(root, f)
            print(f'  {p} ({os.path.getsize(p)/1024/1024:.1f} MB)')
            found = True
if not found: print('  ❌ Nessun APK')

# www/
print('\n──── www/ ────')
www = os.path.join(PROJ,'www') if os.path.exists(PROJ) else None
if www and os.path.exists(www):
    for f in sorted(os.listdir(www)):
        sz = os.path.getsize(os.path.join(www,f))
        print(f'  {f}  ({sz/1024:.0f} KB)')
else:
    print('  ❌ www/ non trovato — esegui Cella 3')

# config.xml
print('\n──── config.xml ────')
cfgp = os.path.join(PROJ,'config.xml')
if os.path.exists(cfgp):
    import re
    cfg = open(cfgp).read()
    for tag in ['id','version','name']:
        m = re.search(rf'{tag}=["\']([^"\']+)["\']', cfg)
        if m: print(f'  {tag}: {m.group(1)}')
else:
    print('  ❌ config.xml non trovato — esegui Cella 4')

# Gradle log
print('\n──── Ultimi errori Gradle ────')
logs = glob.glob(f'{PROJ}/platforms/android/**/*.log', recursive=True)
if logs:
    for lf in logs[-2:]:
        print(open(lf).read()[-2000:])
else:
    # Cerca in build output
    build_logs = glob.glob(f'{PROJ}/platforms/android/**/build/*.txt', recursive=True)
    if build_logs:
        print(open(build_logs[-1]).read()[-2000:])
    else:
        print('  Nessun log trovato')

print('\n' + '═'*60)
```
